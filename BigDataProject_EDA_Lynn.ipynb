## 1. Check missing data by column

```python
def count_missing_value_by_col(dataframe):
    missing_count = dataframe.isnull().sum()
    missing_value_df = pd.concat([missing_count, missing_count/dataframe.shape[0] * 100], axis=1)
    missing_value_df.columns = ['row_count', 'percent']
    return missing_value_df
```

```python
missing_summary = count_missing_value_by_col(df); missing_summary
```

## 2. Unique value by column

```python
def count_unique_value_by_col(dataframe):
    res=[]
    datatype = dataframe.dtypes
    cols = dataframe.columns
    for i in range(len(cols)):
        if datatype[i] != 'object':
            df[cols[i]].astype('str')
        res.append(len(df[cols[i]].unique()))
    res_df = pd.DataFrame(res, index=cols)
    res_df.columns = ['Count_of_unique_val']
    return res_df
```

```python
unique_val_summary =count_unique_value_by_col(df); unique_val_summary
```
